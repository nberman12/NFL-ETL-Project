```python
import sqlite3
import csv
```

```python
conn = sqlite3.connect("NFL_ETL.sqlite")
cur = conn.cursor()
```

```python
cur.execute('DROP TABLE IF EXISTS team')
cur.execute('''
CREATE TABLE "team"(
    
    "TM" TEXT PRIMARY KEY,
    "Team" TEXT
)
''')

fname = "Resources/team_df.csv"

with open(fname) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter= ',')
    for row in csv_reader:
        if row[0] != '':
            print(row)
            TM = row[1]
            Team = row[2]
            
            cur.execute('''INSERT INTO team(TM,Team)
                VALUES (?,?)''',(TM,Team))
            conn.commit
```

```python
cur.execute('DROP TABLE IF EXISTS stadium')
cur.execute('''
CREATE TABLE "stadium"(
    
"stad_id" INT PRIMARY KEY,
"Name" TEXT,
"Capacity" INT,
"City" TEXT,
"State" TEXT,
"Surface" TEXT,
"Roof_Type" TEXT,
"Opened" INT


)
''')

fname = "Resources/stadium_df.csv"

with open(fname) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter= ',')
    for row in csv_reader:
        if row[0] != '':
            print(row)
            stad_id = row[1]
            Name = row[2]
            Capacity = row[3]
            City = row[4]
            State = row[5]
            Surface = row[6]
            Roof_Type = row[7]
            Opened = row[8]


            
            cur.execute('''INSERT INTO stadium(stad_id,Name,Capacity,City,State,Surface,Roof_Type,Opened)
                VALUES (?,?,?,?,?,?,?,?)''',(stad_id,Name,Capacity,City,State,Surface,Roof_Type,Opened))
            conn.commit
```

```python
cur.execute('DROP TABLE IF EXISTS owners')
cur.execute('''
CREATE TABLE "owners"(
    
"TeamID" TEXT,
"Owner" TEXT,
"Year_Acquired" TEXT,
"stad_id" TEXT,
FOREIGN KEY(TeamID) REFERENCES team(TM),
FOREIGN KEY(stad_id) REFERENCES stadium(stad_id)


)
''')

fname = "Resources/owner.csv"

with open(fname) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter= ',')
    for row in csv_reader:
        if row[0] != '':
            print(row)
            TeamID = row[1]
            Owner = row[2]
            Year_Acquired = row[3]
            stad_id = row[4]

            
            cur.execute('''INSERT INTO owners(TeamID,Owner,Year_Acquired,stad_id)
                VALUES (?,?,?,?)''',(TeamID,Owner,Year_Acquired,stad_id))
            conn.commit
```

```python
cur.execute('DROP TABLE IF EXISTS players')
cur.execute('''
CREATE TABLE "players"(
    
    "PlayerID" TEXT PRIMARY KEY,
    "POS" TEXT,
    "Age" REAL,
    "HT" TEXT,
    "WT" TEXT,
    "Exp" TEXT,
    "College" TEXT,
    "FullName" TEXT,
    "First_Name" TEXT,
    "Last_Name" TEXT,
    "TeamID" TEXT,
    FOREIGN KEY(TeamID) REFERENCES team(TM)
)
''')

fname = "Resources/players_df.csv"

with open(fname) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter= ',')
    for row in csv_reader:
        if row[0] != '':
            print(row)
            PlayerID = row[1]
            POS = row[2]
            Age = row[3]
            HT = row[4]
            WT = row[5]
            Exp = row[6]
            College = row[7]
            FullName = row[8]
            First_Name = row[9]
            Last_Name = row[10]
            TeamID = row[11] 
            cur.execute('''INSERT INTO players(PlayerID,POS,Age,HT,WT,Exp,College,FullName,First_Name,Last_Name,TeamID)
                VALUES (?,?,?,?,?,?,?,?,?,?,?)''',(PlayerID,POS,Age,HT,WT,Exp,College,FullName,First_Name,Last_Name,TeamID))
            conn.commit
```
