```python
from bs4 import *
import requests
import re
import pandas as pd
```

```python
df = pd.read_csv("team_links.csv")
```

```python
master_offense = pd.DataFrame({})
master_defense = pd.DataFrame({})
master_special = pd.DataFrame({})


for i in range(0,len(df.Team)):
#for i in range(0,2):
    url = df.roster_links[i]
    table=pd.read_html(url)
    
    offense_table = table[0][["Name","POS","Age","HT","WT","Exp","College"]]
    offense_table["team"]=df.Team[i]
    defense_table = table[1][["Name","POS","Age","HT","WT","Exp","College"]]
    defense_table["team"]=df.Team[i]
    special_table = table[2][["Name","POS","Age","HT","WT","Exp","College"]]
    special_table["team"]=df.Team[i]
    

    #merge all the tables into master DF
    master_offense=pd.concat([master_offense,offense_table])
    master_defense=pd.concat([master_defense,defense_table])
    master_special=pd.concat([master_special,special_table])
   
```

```python
master_offense=master_offense.reset_index(drop=True)
master_defense=master_defense.reset_index(drop=True)
master_special=master_special.reset_index(drop=True)

master_offense.to_csv("roster/offense.csv")
master_defense.to_csv("roster/defense.csv")
master_special.to_csv("roster/special.csv")

```
