```python
import sqlite3
import csv
```

```python
conn = sqlite3.connect("NFL_ETL.sqlite")
cur = conn.cursor()
```

```python
cur.execute('DROP TABLE IF EXISTS players')
cur.execute('''
CREATE TABLE "players"(
    
    "PlayerID" TEXT,
    "POS" TEXT,
    "Age" REAL,
    "HT" TEXT,
    "WT" TEXT,
    "Exp" TEXT,
    "College" TEXT,
    "FullName" TEXT,
    "First_Name" TEXT,
    "Last_Name" TEXT,
    "TeamID" TEXT
)
''')

fname = "temp/players_df.csv"

with open(fname) as csv_file:
    csv_reader = csv.reader(csv_file, delimiter= ',')
    for row in csv_reader:
        if row[0] != '':
            print(row)
            PlayerID = row[1]
            POS = row[2]
            Age = row[3]
            HT = row[4]
            WT = row[5]
            Exp = row[6]
            College = row[7]
            FullName = row[8]
            First_Name = row[9]
            Last_Name = row[10]
            TeamID = row[11] 
            cur.execute('''INSERT INTO players(PlayerID,POS,Age,HT,WT,Exp,College,FullName,First_Name,Last_Name,TeamID)
                VALUES (?,?,?,?,?,?,?,?,?,?,?)''',(PlayerID,POS,Age,HT,WT,Exp,College,FullName,First_Name,Last_Name,TeamID))
            conn.commit
```
